```python
# clean out special characters
def col_cleaner(df):
    for col in df.columns:
        df[col] = df[col].apply(lambda x: re.sub("[^0-9a-zA-Z-@]+", " ", str(x)).lower())
        try:
            df[col] = pd.to_numeric(df[col])
        except:
            pass
    return df
```
