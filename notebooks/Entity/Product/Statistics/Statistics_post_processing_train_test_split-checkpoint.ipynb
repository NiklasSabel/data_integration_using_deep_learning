# Statistics for filtered tables after post processing

```python
all_filtered_tables_df = pd.read_json(os.path.join(train_test_all_filtered_path,'train_test_all_filtered_tables.json.gz'), compression='gzip', lines=True)
all_filtered_tables_df.head()
```

```python
overview_cluster_tablecounts = all_filtered_tables_df['cluster_id'].value_counts().to_frame().reset_index().rename(columns={'index':'cluster_id', 'cluster_id':'counts'})
overview_cluster_tablecounts['cluster_id'] = overview_cluster_tablecounts['cluster_id'].apply(lambda x: str(x))
overview_cluster_tablecounts
```

```python
overview_cluster_tablecounts.describe()
```

```python
overview_cluster_tablecounts.loc[overview_cluster_tablecounts['counts']<= 10,'size'] = '<= 10'
overview_cluster_tablecounts.loc[(overview_cluster_tablecounts['counts']> 10) & (overview_cluster_tablecounts['counts']<= 20),'size'] = '11-20'
overview_cluster_tablecounts.loc[(overview_cluster_tablecounts['counts']> 20) & (overview_cluster_tablecounts['counts']<= 50),'size'] = '21-50'
overview_cluster_tablecounts.loc[(overview_cluster_tablecounts['counts']> 50),'size'] = '> 50'
overview_cluster_tablecounts
```

```python
overview_cluster_tablesizes = overview_cluster_tablecounts['size'].value_counts().to_frame().reset_index().rename(columns={'index':'size', 'size':'counts'})
overview_cluster_tablesizes = overview_cluster_tablesizes.reindex([1,0,2,3])
overview_cluster_tablesizes
```

```python
plt.bar(list(overview_cluster_tablesizes['size']), list(overview_cluster_tablesizes['counts']))
plt.title('Cluster sizes - Product')
plt.xlabel('# tables')
plt.ylabel('counts')
plt.savefig(os.path.join(visualization_path, 'After Post Processing', 'Cluster sizes.jpg'))
plt.show()
```

```python
tables_overview_df = pd.read_excel(os.path.join(train_test_output_path, 'large', 'Test_Data_Check.xlsx')).drop(columns='Unnamed: 0')
tables_overview_df
```

```python
distribution_split_df = tables_overview_df.groupby(['cluster_id', 'dataset'])['dataset'].size().reset_index(name='single_count')
distribution_split_df
```

```python
distribution_not_split_df = tables_overview_df.groupby(['cluster_id'])['dataset'].size().reset_index(name='together_count')
distribution_not_split_df
```

```python
distribution_all_df = pd.merge(distribution_not_split_df, distribution_split_df, left_on='cluster_id', right_on='cluster_id', how='left')
distribution_all_df['ratio'] = distribution_all_df['single_count'] / distribution_all_df['together_count']
distribution_all_df['cluster_id'] = distribution_all_df['cluster_id'].apply(lambda x: str(x))
distribution_all_df
```

```python
distribution_test_df = distribution_all_df[distribution_all_df['dataset']=='test']
distribution_train_df = distribution_all_df[distribution_all_df['dataset']=='train']
distribution_val_df = distribution_all_df[distribution_all_df['dataset']=='val']
```

```python
distribution_test_statistics = distribution_test_df.describe()
distribution_test_statistics
```

```python
test_mean = round(distribution_test_statistics.loc['mean', 'ratio'],4)
test_median = round(distribution_test_statistics.loc['50%', 'ratio'],4)
```

```python
plt.figure(figsize=(15,10))
sns.distplot(distribution_test_df['ratio'])
plt.title('Distribution of test ratio - Product')
plt.savefig(os.path.join(visualization_path, 'Train test split', 'Distribution plot_ratio test set large.jpg'))
```

```python
plt.figure(figsize=(15,10))
graph = sns.barplot(data = distribution_test_df, x='cluster_id', y='ratio', color='blue')
graph.axhline(y=test_mean, color='green', label='mean ratio = {}'.format(test_mean))
graph.axhline(y=test_median, color='red', label='median ratio = {}'.format(test_median))
graph.legend(loc='upper right')
graph.set(xlabel='cluster', ylabel='test size ratio', title='Clusters with test size ratio - Product')
graph.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False) 
plt.savefig(os.path.join(visualization_path, 'Train test split', 'Clusters with test size ratio.jpg'))
```
